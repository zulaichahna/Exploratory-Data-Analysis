```python
cats = ['Gender',
        'Occupation',
        'BMI Category',
        'Blood Pressure',
        'Sleep Disorder']
nums = ['Age', 
        'Sleep Duration', 
        'Quality of Sleep', 
        'Physical Activity Level', 
        'Stress Level', 
        'Heart Rate',
        'Daily Steps']
```

```python
plt.figure(figsize=(12, 4))
for i in range(0, len(nums)):
    plt.subplot(1, len(nums), i+1)
    sns.boxplot(y=df[nums[i]], orient='v')
    plt.tight_layout()
```

```python
sns.set_style("darkgrid")
                     
plt.figure(figsize=(14, len(nums) * 3))
                     
for idx, feature in enumerate(nums, 1):
    plt.subplot(len(nums), 2, idx)
    sns.histplot(df[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(df[feature].skew(), 2)}")
                     
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

for i in range(0, len(cats)):
    plt.subplot(2, 3, i+1)
    sns.countplot(x=df[cats[i]])
    plt.xlabel(cats[i])
    plt.xticks(rotation=20)
    plt.tight_layout()
```

```python
plt.figure(figsize=(8, 5))
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), cmap='Blues', annot=True, fmt='.2f')
```

```python
sns.regplot(x="Sleep Duration", y="Quality of Sleep", data=df)
plt.ylim(0,)
plt.show()
print(df[["Sleep Duration", "Quality of Sleep"]].corr())
```
